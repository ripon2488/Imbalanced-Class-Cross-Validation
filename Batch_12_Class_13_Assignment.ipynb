# Task: 2 # SMOTETomek Sampling

```python
#task:2 Do SMOTETomek sampling on given dataset and Create ML model and evaluate model by cross validation again
```

```python
from imblearn.combine import SMOTETomek
```

```python
smot=SMOTETomek(random_state=42)
```

```python
y.value_counts() # before sampling
```

```python
X_smot,Y_smot = smot.fit_resample(x,y)
```

```python
X_smot.shape
```

```python
Y_smot.shape
```

```python
Y_smot.value_counts()
```

```python
X_smot.head()
```

```python
from imblearn.under_sampling import NearMiss
```

```python
# https://imbalanced-learn.org/dev/references/generated/imblearn.under_sampling.NearMiss.html
```

```python
nearmis=NearMiss()
```

```python
X_nearmis,Y_nearmis=nearmis.fit_resample(x,y)
```

```python
X_nearmis.shape
```

```python
X_nearmis.head()
```
